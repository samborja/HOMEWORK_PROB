```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

tesla = pd.read_csv("C:/Users/JAIRO BORJA/Desktop/UNIVERSIDAD/2SEMESTRE/INTELIGENCIA/TSLA.csv")
tesla.head(1)
```

```python
tesla1 = tesla.iloc[0:571]
tesla1['Close'].mean()
```

```python
x = tesla1['Date']
y = tesla1['Close']

plt.plot(x,y)
plt.xlabel('Date')
plt.ylabel('Close')
plt.title('Date Vs. Close')
plt.show()
```

```python
Omax = tesla1['Open'].max()
Omin = tesla1['Open'].min()

tesla1['OpenNorm'] = tesla1['Open'].apply(lambda x: (x - Omin)/(Omax - Omin))

x1 = tesla1['Date']
y1 = tesla1['OpenNorm']


plt.xlabel('Fecha')
plt.title('Date Vs. OpenNorm.')
plt.xticks([0, 285, 570])
plt.plot(x1, y1, color = 'r', label = 'Open')
plt.show()
```

```python
tesla1['SMA12'] = tesla1['OpenNorm'].rolling(12).mean()
tesla1['SMA15'] = tesla1['OpenNorm'].rolling(15).mean()
tesla1['SMA30'] = tesla1['OpenNorm'].rolling(30).mean()

x1 = tesla1['Date']
y1 = tesla1['OpenNorm']
y2 = tesla1['SMA12']
y3 = tesla1['SMA15']
y4 = tesla1['SMA30']

plt.xlabel('Fecha')
plt.title('SMA (12, 15, 30)')
plt.xticks([0, 285, 570])
plt.plot(x1, y1, color = 'r', label = 'Open')
plt.plot(x1, y2, color = 'b', label = 'SMA12')
plt.plot(x1, y3, color = 'y', label = 'SMA15')
plt.plot(x1, y4, color = 'g', label = 'SMA30')
plt.show()
```
